
Make the retail return dataset Tableau-ready.
Adds a numeric return flag, binned helper dimensions, and an order-value
measure so charts can be built in Tableau without manual binning.

```python

df = pd.read_csv("/Users/venugopalnaidupotturi/Downloads/Retail_Return_Behaviour_Cleaned.csv")

 
    # 1. Numeric return flag (backbone of the Return Rate calc in Tableau)
df["returned_flag"] = (df["returned"] == "Yes").astype(int)
 
    # 2. Age groups
df["age_group"] = pd.cut(df["customer_age"],bins=[17, 25, 35, 45, 55, 100],labels=["18-25", "26-35", "36-45", "46-55", "56+"], )
 
    # 3. Price bands
df["price_band"] = pd.cut(df["product_price"],bins=[0, 25, 50, 100, 200, 500, 1e9],labels=["<$25", "$25-50", "$50-100", "$100-200", "$200-500", "$500+"],
)
 
    # 4. Delivery delay buckets
df["delivery_delay_bucket"] = pd.cut(df["delivery_delay_days"],bins=[-2, 0, 2, 5, 100],labels=["On-time/Early", "1-2 days late", "3-5 days late", "5+ days late"],
)
 
    # 5. Customer rating bands
df["rating_band"] = pd.cut(df["customer_rating"],bins=[0, 2, 3, 4, 5],labels=["1-2 (Poor)", "2-3 (Fair)", "3-4 (Good)", "4-5 (Great)"],
    )
 
    # 6. Past return rate bands
df["past_return_band"] = pd.cut(df["past_return_rate"],bins=[-0.01, 0.1, 0.2, 0.3, 1],labels=["Low (<10%)", "Med (10-20%)", "High (20-30%)", "V.High (30%+)"],
    )
 
    # 7. Order value measure (price * quantity)
df["order_value"] = (df["product_price"] * df["order_quantity"]).round(2)
 
    
```
